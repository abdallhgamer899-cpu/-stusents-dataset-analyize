```python
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output

# إعداد التنسيق الجمالي (CSS)
custom_css = """
<style>
    .gift-container {
        text-align: center;
        background: linear-gradient(135deg, #fce4ec 0%, #f06292 100%);
        padding: 40px;
        border-radius: 20px;
        box-shadow: 0 10px 20px rgba(0,0,0,0.1);
        font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
    }
    .gift-text { color: #880e4f; font-size: 24px; font-weight: bold; margin-bottom: 20px; }
    .gift-emoji { font-size: 50px; animation: bounce 1s infinite; }
    @keyframes bounce {
        0%, 100% { transform: translateY(0); }
        50% { transform: translateY(-15px); }
    }
    .reveal-btn {
        background-color: #c2185b !important;
        color: white !important;
        font-weight: bold !important;
        border-radius: 10px !important;
    }
</style>
"""

# المحتوى الذي سيظهر عند النقر
gift_content = """
<div class="gift-container">
    <div class="gift-emoji">🎁✨</div>
    <div class="gift-text">Happy Birthday, Farah! 🎂</div>
    <p style="color: #ad1457;">Wishing you a day full of joy and code that runs without errors.</p>
</div>
"""

# إنشاء الزر
button = widgets.Button(description="Open Your Gift 🎁", button_style='info')
button.add_class("reveal-btn")
output = widgets.Output()

def on_button_clicked(b):
    with output:
        clear_output()
        display(HTML(gift_content))

button.on_click(on_button_clicked)

# العرض
display(HTML(custom_css))
display(button, output)
```
